```python
df = pd.read_csv('105.csv',lineterminator='\n')
df.head(24)
```

```python
df = pd.read_csv('106.csv',lineterminator='\n')

df.head(50)
```

```python
df_1 = pd.concat([pd.read_csv('105.csv')])
df_2 = pd.concat([pd.read_csv('106.csv')])
df = pd.concat([df_1,df_2])
df
```

1. Which city had the highest number of public preschools in 106?

```python

# Read the data from the file
data = df

# Sort the data by the "公立園數" column in descending order
sorted_data = data.sort_values("公立園數", ascending=False)

# Get the city with the highest number of public preschools
city = sorted_data.iloc[0]["縣市別"]

print("The city with the highest number of public preschools is", city)
```

2. What was the total number of preschools in Yunlin County in 105?

```python
# Filter the rows for Yunlin County and select the "園數總數" column
yunlin_preschools = df[df['縣市別'] == '雲林縣']['園數總數']

# Print the total number of preschools in Yunlin County
print('Total number of preschools in Yunlin County in 105 is', yunlin_preschools.values[0])
```

3. How many preschools had specialized classes in Hualien County in 105?

```python
data = df
# Select only the rows with year 105
data = df[df['年度'] == 105]

hualien_data = df[df['縣市別'] == '花蓮縣']
specialized_count = hualien_data['公立專設園數'].sum()


print(f"Total number of specialized public preschools in Hualien County in 105 is {specialized_count}")
```

4. Which county had the highest number of preschool teachers in 105?

```python

# sort the data by the "教師總數" column in descending order
sorted_df = df.sort_values("教師總數", ascending=False)

# print the county with the highest number of preschool teachers
print(sorted_df.iloc[0]["縣市別"],"has the highest number of preschool teacher.")
```

5. What was the total number of preschool students in Taipei City in 105?

```python
data = df[df['年度'] == 105]
# filter the data to only include Taipei City
taipei_df = data[data["縣市別"] == "臺北市"]

# sum up the "幼生總數" column
total_students = taipei_df["幼生總數"].sum()

print("Total number of preschool students in Taipei City in 105 is", total_students)
```

6. Which city had the lowest number of public preschools in 106?

```python

data = df

min_public_preschools = data['公立園數'].min()
city_with_min_public_preschools = df.loc[data['公立園數'] == min_public_preschools, '縣市別'].iloc[0]

print(city_with_min_public_preschools,"had the lowest number of public preschools in 106")
```

7. How many preschools had assistant teachers in Taoyuan City in 105?

```python
data = df

assistant_teachers = df.loc[data["縣市別"] == "桃園市", "助理教保員總數"].values[0]

print("The number of preschools with assistant teachers in Taoyuan City in 2016 is", assistant_teachers)
```

8. What was the total number of preschool service personnel in Tainan City in 105?

```python
data = df
# filter the dataframe to only include rows for Tainan City
tainan_df = df[data['縣市別'] == '臺南市']

# select the "服務人員總數" column and sum the values
total_service_personnel = tainan_df['服務人員總數'].sum()

print("Total number of preschool service personnel in Tainan City in 105 is", total_service_personnel)
```

9. Which county had the highest number of preschool students in 105?

```python
data = df

# Group the data by county and sum the number of preschool students in each county
county_totals = data.groupby("縣市別")["幼生總數"].sum()

# Sort the data in descending order
county_totals = county_totals.sort_values(ascending=False)

# Print the county with the highest number of preschool students
print("County with highest number of preschool students in 105 is", county_totals.index[0])

```

10. How many preschools had public teachers in Kaohsiung City in 105?

```python
data = df
# Filter the data to only include rows where the "縣市別" column contains "高雄市"
df_kh = data[data["縣市別"].str.contains("高雄市")]

# Sum up the values in the "公立園數" column
num_public_preschools = df_kh["公立園數"].sum()

# Print the result
print(num_public_preschools,"preschools had public teachers in Kaohsiung City in 105.")
```
